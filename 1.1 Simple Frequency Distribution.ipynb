# Simple Frequency Distribution

```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python
#Initialize Tokenizer
tokenizer=TreebankWordTokenizer()
```

```python
#Sample Text
text="Natural Lanuage Processing (NLP) with python is fun and educational. Python simplifies NLP tasks."
```

```python
#Tokenizer Text
tokens=tokenizer.tokenize(text)
```

```python
#Print tokens
print("Tokens:", tokens)
```

```python
#Frequency Distribution
fdist=FreqDist(tokens)
```

```python
#Print Frequencies
print("\n Word Frequency Distribution:")
for word,freq in fdist.items():
    print(f"{word}:{freq}")
```

```python
#Plot Frequency Distribution
fdist.plot(title='Simple Frequency Distribution Plot')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.hist(tokens,bins=15)
plt.show()
```
