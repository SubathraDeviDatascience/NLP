```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python
#Download stopwords set (if not already downloaded)
nltk.download('stopwords')
```

```python
#Initialize tokenizer and stopwords list
tokenizer=TreebankWordTokenizer()
stop_words=set(stopwords.words('english'))
```

```python
#Sample Text
text="Natural language processing with pyhon is interesting and useful. Python makes NLP tasks easier and more efficient."
```

```python
#Tokenize the text
tokens=tokenizer.tokenize(text)
```

```python
#Remove Stopwords
filtered_tokens=[word for word in tokens if word.lower() not in stop_words and word.isalpha()]
```

```python
#Print filtered tokens
print("Filtered Tokens (without stopwords): ", filtered_tokens)
```

```python
#Frequency Distribution
fdist=FreqDist(filtered_tokens)
```

```python
#Print frequency distributions
print("\n Word Frequency Distribution (without stopwords):")
for word, freq in fdist.items():
    print(f"{word}:{freq}")
```

```python
#Plot frequency distribution
fdist.plot(title='Word Frequency Distribution (without stopwords)')
plt.show()   
```

```python
plt.figure(figsize=(15,8))
plt.hist(filtered_tokens,bins=15)
plt.show()
```
